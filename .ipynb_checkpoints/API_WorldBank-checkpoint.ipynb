```python
# Étape 1 : remettre les années en ligne (stack)
df_reset = df.stack()                        # transforme les colonnes YR2010...YR2023 en lignes
df_reset = df_reset.reset_index()            # transforme l'index en colonnes normales
df_reset.columns = ['pays', 'indicateur', 'annee', 'valeur']

print(df_reset.head(10))  # vérifie la structure
```

```python
# Étape 2 : pour chaque pays + indicateur, garder uniquement la valeur la plus récente non nulle
df_recent = (
    df_reset
    .dropna(subset=['valeur'])                        # supprimer les lignes sans valeur
    .sort_values('annee', ascending=False)            # trier par année décroissante
    #.groupby(['pays', 'indicateur'])                  # grouper par pays + indicateur
    .first()                                          # garder la première ligne = la plus récente
    .reset_index()
)

df_recent.head(10)
```

```python
# Étape 3 : pivoter pour avoir une colonne par indicateur
df_final = df_recent.pivot(index = 'pays', columns='indicateur', values='valeur')

# Étape 4 : renommer les colonnes avec les noms lisibles
df_final = df_final.rename(columns = indicateurs)

# Étape 5 : remettre 'pays' comme colonne normale
df_final = df_final.reset_index()
df_final

# Sauvegarde
#df_final.to_csv("donnees_touriste.csv", index=False)
#print("💾 Fichiers sauvegardés !")
```
