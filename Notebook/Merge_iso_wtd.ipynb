# Importations dataframes

# Mise à jour des colonnes
Je les mets toutes en minuscules pour pouvoir garentir un merge de _qualité_

```python
df_lexique["pays"] = df_lexique["pays"].str.lower()
df_wtd["country"] = df_wtd["country"].str.lower()
```

# Merge

```python
df_merge = pd.merge(df_lexique, df_wtd, left_on="pays", right_on="country", how="inner")
df_merge
```

```python
df_merge.info()
```

# Extraction des données

```python
worldtraveldestination_iso = df_merge
```

```python
output_dir = "../DATA/processed"
os.makedirs(output_dir, exist_ok=True)

# Base de données déscriptives
worldtraveldestination_iso.to_csv(os.path.join(output_dir, "WorldTravel_ISO.csv"))
```
